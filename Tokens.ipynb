```python
import nltk
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def text_process(text):
    """
    Take in a string of text, then perform the following:
    1. Remove all punctuation
    2. Remove all stop words
    3. Return a list of the cleaned text
    """
    # check characters to see if they are in punctuations
    nopunc_lst = [char for char in text if char not in string.punctuation]
    
    # join the characters again to form a string
    nopunc = ''.join(nopunc_lst)
    
    # remove stop words
    cleaned_text_lst = [word.lower() for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    
    return cleaned_text_lst
```
